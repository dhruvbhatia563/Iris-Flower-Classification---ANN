# **Iris Dataset** -  ANN
------
### **Classify the flower is Setosa or not using Sepal and Petal (Length and Width)**

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
import sklearn
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

----------

```python
iris #sepal-length, sepal-width, petal-length, petal-width
```

```python
iris.target #setosa, versicolor,virginica
```

```python
setosa_y = (iris.target==0).astype(np.int)
setosa_y
```

```python
x = iris.data
x[0:9]
```

```python
df = pd.DataFrame(x, columns = ['Sepal Length','Sepal Width','Petal Length','Petal Width'])
df.head()
```

```python
len(setosa_y), len(df)
```

```python
df['Setosa'] = 0
```

```python
for i in range(len(df)):
    df['Setosa'][i] = setosa_y[i]
    
df.head()    
```

-----------------

```python
y = df['Setosa']
x = df.drop('Setosa',axis=1)
```

```python
x.head(2)
```

```python
y.head(2)
```

```python
df.head(2)
```

```python
x.corr()
```

```python
new_x = df[['Sepal Width', 'Petal Length']]
new_x.head()
```

```python
y.head()
```

```python
new_x.shape
```

```python
y.shape
```

------------
#### **Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(new_x,y,train_size=0.7,random_state=42)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

-------
#### **Training the dataset - ANN**

```python
from sklearn.linear_model import Perceptron
```

```python
model = Perceptron(random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
model.classes_
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_train_pred = model.predict(x_train)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
print(classification_report(y_train,y_train_pred))
```

```python
conf_matrix = pd.DataFrame(confusion_matrix(y_train,y_train_pred),columns=['Not_Setosa','Setosa'])
conf_matrix
```

```python
print(f'accuracy of training: {accuracy_score(y_train,y_train_pred)}')
```

-------
#### **Testing the dataset - ANN**

```python
x_test.shape,y_test.shape
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
y_test_pred = model.predict(x_test)
```

```python
print(classification_report(y_test,y_test_pred))
```

```python
conf_matrix = pd.DataFrame(confusion_matrix(y_test,y_test_pred),columns=['Not_Setosa','Setosa'])
conf_matrix
```

```python
print(f'accuracy of test: {accuracy_score(y_test,y_test_pred)}')
```
